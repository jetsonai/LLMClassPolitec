```python
from konlpy.tag import Okt
okt = Okt()
text = "자연어 처리는 인공지능의 한 분야로서, 인간의 언어를 이해하고 처리합니다."
tokens = okt.morphs(text)
stop_words = ['은', '는', '이', '가', '을', '를', '에', '의', '도', '으로서', '하고', '합니다', ',', '.']
filtered_tokens = [word for word in tokens if word not in stop_words]
print("불용어 제거 후 토큰들: ", filtered_tokens)
```

```python
from konlpy.tag import Okt

okt = Okt()
text = "열심히 공부하고 있는 학생들은 좋은 성적을 받을 것입니다."

morphs = okt.morphs(text, stem=True)
print("어간 추출 결과:", morphs)
morphs = okt.morphs(text, stem=False)
print("어간 추출 없이 형태소 분석 결과:", morphs)
```

```python
sentences = [
    "나는 자연어 처리를 공부합니다.",
    "자연어 처리는 재미있습니다.",
    "딥러닝을 이용한 자연어 처리"
]
import re
def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', text)
    tokens = text.split()
    return tokens
tokenized_sentences = [preprocess_text(sentence) for sentence in sentences]
print(tokenized_sentences)
vocab = set()
for tokens in tokenized_sentences:
    vocab.update(tokens)
vocab = sorted(vocab)
print(vocab)
word_to_index = {word: idx for idx, word in enumerate(vocab)}
print(word_to_index)
import numpy as np
def one_hot_encode(tokens, word_to_index):
    vector = np.zeros(len(word_to_index))
    for token in tokens:
        index = word_to_index[token]
        vector[index] = 1
    return vector
one_hot_vectors = [one_hot_encode(tokens, word_to_index) for tokens in tokenized_sentences]
for idx, vector in enumerate(one_hot_vectors):
    print(f"문장 {idx + 1}: {vector}")
```

```python
sentences = [
    "나는 자연어 처리를 공부합니다.",
    "자연어 처리는 재미있습니다.",
    "딥러닝을 이용한 자연어 처리"
]
from konlpy.tag import Okt
okt = Okt()
def preprocess_text(text):
    morphs = okt.morphs(text, stem='True')
    stop_words = ['은', '는', '이', '가', '을', '를', '에', '의', '도', '으로서', '하고', '합니다', ',', '.']
    tokens = [word for word in morphs if word not in stop_words]
    return tokens
tokenized_sentences = [preprocess_text(sentence) for sentence in sentences]
print(tokenized_sentences)
vocab = set()
for tokens in tokenized_sentences:
    vocab.update(tokens)
vocab = sorted(vocab)
print(vocab)
word_to_index = {word: idx for idx, word in enumerate(vocab)}
print(word_to_index)
import numpy as np
def one_hot_encode(tokens, word_to_index):
    vector = np.zeros(len(word_to_index))
    for token in tokens:
        index = word_to_index[token]
        vector[index] = 1
    return vector
one_hot_vectors = [one_hot_encode(tokens, word_to_index) for tokens in tokenized_sentences]
for idx, vector in enumerate(one_hot_vectors):
    print(f"문장 {idx + 1}: {vector}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
sentences = [
    "나는 자연어 처리를 공부합니다.",
    "자연어 처리는 재미있습니다.",
    "딥러닝을 이용한 자연어 처리"
]
vectorizer = CountVectorizer(binary=True)
X = vectorizer.fit_transform(sentences)
print("어휘 사전: ", vectorizer.get_feature_names_out())
print("희소 행렬 형태:", X)
print("밀집 행렬 형태:", X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents = [
    "자연어 처리는 재미있다",
    "자연어 처리는 어려운 만큼 재미있다",
    "나는 자연어 처리를 공부한다",
    "데이터 분석도 공부한다"
]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
print("단어 목록: ", vectorizer.get_feature_names_out())
print("TF-IDF 행렬:", tfidf_matrix.toarray())
```
